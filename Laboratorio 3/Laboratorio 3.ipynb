Algoritmo Voraces que resuelve el problema del cambio CMP.
Se van tomando las denominaciones mas grandes que puedan ser tomadas dentro del rango a pagar.

```python
def changeCMP(d,N):
  #Ordenar las denominaciones de mayor a menor
  d.sort(reverse=True)

  #Imprimir las denominaciones ordenadas
  print("Denominaciones ordenadas: ")
  print(d)

  monedas = []

  while N !=0:
    for i in range(0,len(d)):
      if d[i] <= N:
        monedas.append(d[i])
        N = N-d[i]
        break

  #Imprimir la cantidad de monedas de cada denominacion
  print("Cantidad de monedas de cada denominacion: ")
  for i in range(0,len(d)):
    print(d[i],":",monedas.count(d[i]))

d = [1,3,4]
N= 6
changeCMP(d,N)
```

Instancia del problema con la solucion optima

```python
d = [1,2,5]
N= 7
changeCMP(d,N)
```

Algoritmo voraz que resuelve el problema de la mochila 0/1. Se toman los valores mas grandes, siempre y cuando haya suficiente peso disponible.

```python

def mochila01(capacidad, pesos, valores):
  #Crear tupla
  items = list(zip(pesos, valores))
  #Ordenar tupla por mayor valor
  items.sort(key=lambda x: x[1], reverse=True)

  #Imprimir tupla ordenada
  print("Tupla ordenada: ")
  print(items)

  itemsTomados = []
  valorTotal = 0


  for i in range(0,len(items)):
    if items[i][0] <= capacidad:
      capacidad = capacidad - items[i][0]
      itemsTomados.append(items[i])
      valorTotal = valorTotal + items[i][1]


  #Imprimir items tomados
  print("Items tomados: ")
  print(itemsTomados)

  #Imprimir valor total
  print("Valor total: ")
  print(valorTotal)





pesos = [10, 20, 30]
valores = [60, 100, 120]
capacidad = 50

mochila01(capacidad, pesos, valores)
```
