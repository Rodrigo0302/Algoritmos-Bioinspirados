Crea una lista con los valores de los dos primeros periodos de la funcion f(t)=sin(2*pi*t/5 + 0.2).
Grafica los puntos (t,f(t))

```python
import matplotlib.pyplot as plt

t = np.linspace(0, 2*5, 300)
f = np.sin(2*np.pi*t/5 + 0.2)
plt.plot(t, f)
plt.show()
```

Encuentra los puntos donde la tangente de f(t) tiene una pendiente aproximada a cero y muestralos en la grafica. Pista: Aproxima la tangente utiizando la secante.

```python
#Pendiente de la secante
slope = np.diff(f)/np.diff(t)
#Encuentra todos los puntos donde la pendiente es cero
index = np.where(np.abs(slope) < 0.1)
plt.plot(t, f)
for i in range(0, len(index)):
    plt.plot(t[index[i]], f[index[i]], 'ro')
plt.show()
```
