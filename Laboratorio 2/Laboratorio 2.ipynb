# **Flores Estopier Rodrigo**
Define m[i,w] como el valor maximo que se puede obtener con un peso menor o igual a w utilizando elementos hasta i(primeros i items).

1.    m[0,w] = 0;     m[i,0] = 0
2.    m[i,w] = m[i-1,w] si wi > w
3.    m[i,w] = max( m[i-1,w], m[i-1, w-wi] + vi )



```python
import numpy as np
v = (2,3,4,5,6)
w = (2,4,5,5,7)
W = 9

#v = (2,3,5,5,6)
#w = (2,4,4,5,7)
#W = 9

N= len(v)

m= np.zeros((N+1, W+1))

for row in range(1, N+1):
    for col in range(1, W+1):
      if w[row-1] > col:
        m[row][col]= m[row-1][col]
      else:
        m[row][col] = max(m[row-1][col], m[row-1][col -w[row-1]]+v[row-1])

m
```

Determinar que objetos hay en  la mochila (debemos recorrer hacia atras la tabla)

```python

i = N
j = W
objetos = []

while i > 0 and j > 0:
  if m[i][j] != m[i-1][j]:
    objetos.append(i-1)
    j = j - w[i-1]
  i = i - 1

print("Objetos seleccionados:", objetos)
```

Solucion Propuesta en clase

```python
itemsInKnapsack = []

row,col = N,W

while row != 0 and col != 0:
  if m[row][col] != m[row-1][col]:
    itemsInKnapsack.append(row-1)
    col = col - w[row-1]
    row = row - 1
  else:
    row = row - 1

m ,itemsInKnapsack
```

**Problema del cambio (Change Making Problem CMP)**
Se debe encontrar el minimo numero de monedas (de cierta denominacion) que se debe dar como cambio de acuerdo a una cantidad de dinero.
###d = [1,2,5]
###N = 7

Denominacion\Cambio N  |1|2|3|4|5|6|7
-----------------------|-|-|-|-|-|-|-----
1                      |1|2|3|4|5|6|7
2                      |1|1|2|2|3|3|4
5                      |1|1|2|2|1|2|2

Ejercicio <br>
d=[1,3,5] <br>
N=6

1.    m[1,N] = N;
2.    m[i,N] = m[i-1,N] if di > N
3.    m[i,N] = min( m[i-1,N], m[i, N-di] + 1 )



```python
d = [1,2,5] #i)deben estar en orden ascendente ii) el primero debe ser 1 iii) cantidad ilimitada de monedas
N = 7

#d = [1,3,5]
#N = 6

W= len(d)

m= np.zeros((W, N))

#Inicializar la primera fila
for col in range(N):
  m[0][col] = col + 1

#Inicializar la primera columna
for row in range(1, W):
  m[row][0] = 1

for row in range(1, W):
  for col in range(1, N):
    if d[row] > col+1:
      m[row][col]= m[row-1][col]
    else:
      m[row][col] = min(m[row-1][col], m[row][col -d[row]]+1)

m
```

Encontrar cuantas monedas de cada denominacion se tomaron en la solucion optima.

```python
# prompt: De la solucion obtenida m, encontrar cuantas monedas de cada denominacion se tomaron

i = W-1
j = N-1

monedas = []

while i > 0 and j > 0:
  if m[i][j] != m[i-1][j]:
    monedas.append(i)
    j = j - d[i]
  else:
    i = i - 1

# Contar la cantidad de monedas de cada denominación
cantidad_monedas = {}
for moneda in monedas:
  if moneda in cantidad_monedas:
    cantidad_monedas[moneda] += 1
  else:
    cantidad_monedas[moneda] = 1

# Imprimir la cantidad de monedas de cada denominación
print("Cantidad de monedas de cada denominación:")
for moneda, cantidad in cantidad_monedas.items():
  print(f"Denominación {d[moneda]}: {cantidad}")
```
